```python
# Step 1: Project Overview and Setup

# Import necessary libraries
import tensorflow as tf
from tensorflow.keras import layers, models, datasets
import numpy as np
import matplotlib.pyplot as plt
from art.attacks.evasion import FastGradientMethod, ProjectedGradientDescent
from art.defences.preprocessor import FeatureSqueezing
from art.defences.trainer import AdversarialTrainer


# Set random seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# CIFAR-10 dataset loading and preprocessing
(x_train, y_train), (x_test, y_test) = datasets.cifar10.load_data()

# Normalize the data to range [0, 1]
x_train, x_test = x_train / 255.0, x_test / 255.0

# Define class names for visualization
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

# Display sample images of vehicles (airplane, automobile, truck, and ship) from the CIFAR-10 dataset
vehicle_classes = [0, 1, 8, 9]  # Indexes for airplane, automobile, ship, and truck classes

plt.figure(figsize=(10, 10))
for i in range(len(x_train)):
    if int(y_train[i]) in vehicle_classes:
        plt.subplot(5, 5, len(plt.gcf().axes) + 1)  # Dynamically adding subplots for vehicle classes
        plt.imshow(x_train[i])
        plt.title(class_names[int(y_train[i])])
        plt.axis('off')

    if len(plt.gcf().axes) == 25:  # Display 25 images only (5x5 grid)
        break

plt.show()
```

## White-Box Attacks:
1. The attacker has complete knowledge of the targeted model, including its architecture, parameters, and training data.
2. White-box attacks are typically more powerful and can achieve higher success rates in crafting adversarial examples.
3. Examples of white-box attacks include the Fast Gradient Sign Method (FGSM), Projected Gradient Descent (PGD), and other optimization-based attacks that rely on model gradients.


## Black-Box Attacks:
1. The attacker has limited or no access to the targeted model's internal details, such as its architecture, parameters, and gradients.
2. Black-box attacks require the attacker to query the model and observe its outputs to generate adversarial examples.
3. These attacks often involve techniques like transferability, where adversarial examples crafted for one model are also effective against another model.
4. Examples of black-box attacks include Zeroth-Order Optimization (ZOO), Query-Efficient Black-Box (QEBB), and other methods that explore model responses without full knowledge of the model.

# Patch Attack

```python
# Step 3: Implementing the Patch Attack

import numpy as np

# Function to apply the patch attack to a single image
def apply_patch_attack(image, patch, target_class, patch_position=(0, 0)):
    # Ensure the patch position is within the image bounds
    img_h, img_w = image.shape[:2]
    patch_h, patch_w = patch.shape[:2]
    x, y = patch_position
    x = np.clip(x, 0, img_w - patch_w)
    y = np.clip(y, 0, img_h - patch_h)

    # Apply the patch to the image
    adversarial_image = np.copy(image)
    adversarial_image[y:y+patch_h, x:x+patch_w] = patch

    return adversarial_image

# Function to generate a patch with a specific color and target class
def generate_patch(patch_size, color=(1.0, 0.0, 0.0), target_class=1):
    patch = np.ones((patch_size, patch_size, 3)) * color
    return patch, target_class

# Find images of vehicles from the test set
vehicle_class_names = ['automobile', 'truck']
vehicle_images = []
for i in range(len(y_test)):
    if class_names[y_test[i][0]] in vehicle_class_names:
        vehicle_images.append(x_test[i])

# Choose the first vehicle image for the patch attack demonstration
sample_vehicle_image = vehicle_images[0]

# Generate a patch with a target class of "truck" (class index 9) and a red color
patch_size = 10
patch, target_class = generate_patch(patch_size, color=(1.0, 0.0, 0.0), target_class=9)

# Apply the patch attack to the vehicle image
patched_vehicle_image = apply_patch_attack(sample_vehicle_image, patch, target_class, patch_position=(5, 5))

# Display the original vehicle image and the adversarial vehicle image
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(sample_vehicle_image)
plt.title("Original Vehicle Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(patched_vehicle_image)
plt.title("Adversarial Vehicle Image (Patch Attack)")
plt.axis('off')

plt.tight_layout()
plt.show()
```

1. The generate_patch() function to allow more flexibility in creating patches. We'll add options to choose the patch color, size, shape, and transparency. Additionally, we'll provide the ability to apply a texture or pattern to the patch.
2. In this code, we modify the generate_patch() function to accept additional parameters: shape, transparency, and texture. The shape parameter allows you to choose between a square or circular patch

```python
import numpy as np

# Function to generate a patch with customizable options
def generate_patch(patch_size, color=(1.0, 0.0, 0.0), shape='square', transparency=1.0, texture=None):
    if shape == 'square':
        patch = np.ones((patch_size, patch_size, 3)) * color
    elif shape == 'circle':
        y, x = np.ogrid[:patch_size, :patch_size]
        center = (patch_size - 1) / 2
        mask = (x - center) ** 2 + (y - center) ** 2 <= (center ** 2)
        patch = np.zeros((patch_size, patch_size, 3))
        patch[mask] = color
    else:
        raise ValueError("Unsupported patch shape. Use 'square' or 'circle'.")

    # Apply transparency to the patch
    patch = patch * transparency + (1.0 - transparency)

    # Apply texture or pattern to the patch
    if texture is not None:
        patch = np.multiply(patch, texture)

    return patch

# Generate and visualize custom patches with different options
patch_size = 10

# Example 1: Red square patch with transparency
patch_1 = generate_patch(patch_size, color=(1.0, 0.0, 0.0), shape='square', transparency=0.5)
# Example 2: Green circular patch with texture
texture = np.random.uniform(0.8, 1.0, size=(patch_size, patch_size, 3))
patch_2 = generate_patch(patch_size, color=(0.0, 1.0, 0.0), shape='circle', transparency=0.8, texture=texture)

# Display custom patches
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(patch_1)
plt.title("Custom Patch 1")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(patch_2)
plt.title("Custom Patch 2")
plt.axis('off')

plt.tight_layout()
plt.show()
```

### 1) SPATIAL SMOOTHING

To apply spatial smoothing defense for the patch attack, we can use a simple image filtering technique such as Gaussian blurring. Gaussian blurring smooths the image by averaging the pixel values in a local neighborhood, which can help reduce the impact of the patch attack.

```python
import cv2

# Function to apply spatial smoothing using Gaussian blurring
def apply_spatial_smoothing(image, kernel_size=5, sigma=1.0):
    smoothed_image = cv2.GaussianBlur(image, (kernel_size, kernel_size), sigma)
    return smoothed_image

# Apply spatial smoothing to the patched vehicle image
smoothed_patched_vehicle_image = apply_spatial_smoothing(patched_vehicle_image)

# Display the original vehicle image, patched vehicle image, and smoothed patched vehicle image
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(sample_vehicle_image)
plt.title("Original Vehicle Image")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(patched_vehicle_image)
plt.title("Adversarial Vehicle Image (Patch Attack)")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(smoothed_patched_vehicle_image)
plt.title("Smoothed Adversarial Vehicle Image")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import cv2

# Function to apply patch attack to an image with spatial smoothing
def apply_patch_attack_with_smoothing(model, x, patch, location=(0, 0), epsilon=0.1, smoothing_kernel_size=5):
    x_adv = np.copy(x)

    # Apply the adversarial patch to the image
    patch_height, patch_width, _ = patch.shape
    x_adv[location[0]:location[0] + patch_height, location[1]:location[1] + patch_width, :] = \
        (1 - epsilon) * x[location[0]:location[0] + patch_height, location[1]:location[1] + patch_width, :] + \
        epsilon * patch

    # Apply spatial smoothing to the adversarial image
    smoothed_image = cv2.GaussianBlur(x_adv, (smoothing_kernel_size, smoothing_kernel_size), 0)

    # Make predictions using the model
    predictions = model.predict(np.expand_dims(smoothed_image, axis=0))

    return smoothed_image, predictions

# Function to apply random patch attack with spatial smoothing
def apply_random_patch_attack_with_smoothing(model, x, patch_size=(30, 30), epsilon=0.1, smoothing_kernel_size=5):
    # Generate a random adversarial patch
    patch = generate_random_patch(patch_size)

    # Randomly select a location to apply the patch
    location = (np.random.randint(0, x.shape[0] - patch_size[0]),
                np.random.randint(0, x.shape[1] - patch_size[1]))

    # Apply patch attack with spatial smoothing
    smoothed_image, predictions = apply_patch_attack_with_smoothing(model, x, patch, location, epsilon, smoothing_kernel_size)

    return smoothed_image, predictions

# Function to apply square patch attack with spatial smoothing
def apply_square_patch_attack_with_smoothing(model, x, patch_size=(30, 30), epsilon=0.1, smoothing_kernel_size=5):
    # Generate a square adversarial patch
    patch = generate_square_patch(patch_size)

    # Randomly select a location to apply the patch
    location = (np.random.randint(0, x.shape[0] - patch_size[0]),
                np.random.randint(0, x.shape[1] - patch_size[1]))

    # Apply patch attack with spatial smoothing
    smoothed_image, predictions = apply_patch_attack_with_smoothing(model, x, patch, location, epsilon, smoothing_kernel_size)

    return smoothed_image, predictions
```

# 2) Randomization

Randomly perturb the patch's position, size, or color during training and inference to make it harder for attackers to generate effective patches.

```python
import random

# Function to apply Randomization to the patch position
def apply_randomization_patch_position(image, patch, max_offset=5):
    img_h, img_w = image.shape[:2]
    patch_h, patch_w = patch.shape[:2]

    # Randomly perturb the patch position
    x_offset = random.randint(-max_offset, max_offset)
    y_offset = random.randint(-max_offset, max_offset)
    x = np.clip(5 + x_offset, 0, img_w - patch_w)
    y = np.clip(5 + y_offset, 0, img_h - patch_h)

    # Apply the patch to the image
    adversarial_image = np.copy(image)
    adversarial_image[y:y+patch_h, x:x+patch_w] = patch

    return adversarial_image

# Example of applying Randomization to the patch position
patched_vehicle_image_random = apply_randomization_patch_position(sample_vehicle_image, patch, max_offset=5)

# Display the original vehicle image and the adversarial vehicle image with Randomization
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(sample_vehicle_image)
plt.title("Original Vehicle Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(patched_vehicle_image_random)
plt.title("Adversarial Vehicle Image (Randomization)")
plt.axis('off')

plt.tight_layout()
plt.show()
```

## FGSM ATTACK

The idea behind FGSM is to add small perturbations to the input data in the direction of the gradient sign of the loss with respect to the input. This causes misclassification with minimal perturbation.

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Function to get model predictions
def get_model_preds(input_image, model):
    logits = model.predict(input_image)
    preds = tf.nn.softmax(logits)
    top_idx = np.argsort(-preds)[0][0]
    top_prob = -np.sort(-preds)[0][0]
    return top_idx, top_prob

# Function to generate adversarial pattern
def generate_adversarial_pattern(input_image, image_label_idx, model, loss_func):
    with tf.GradientTape() as tape:
        tape.watch(input_image)
        prediction = model(input_image)
        loss = loss_func(image_label_idx, prediction)
    gradient = tape.gradient(loss, input_image)
    signed_grad = tf.sign(gradient)
    return signed_grad

# Function to perform FGSM attack
def perform_adversarial_attack_fgsm(input_image, image_label_idx, model, loss_func, eps=0.01):
    # Basic image shaping
    input_image = np.array([input_image])
    tf_img = tf.convert_to_tensor(input_image)
    # Predict class before adversarial attack
    ba_pred_class, ba_pred_prob = get_model_preds(tf_img, model)
    # Generate adversarial image
    adv_pattern = generate_adversarial_pattern(tf_img, image_label_idx, model, loss_func)
    clip_adv_pattern = tf.clip_by_value(adv_pattern, clip_value_min=-eps, clip_value_max=eps)

    perturbed_img = tf_img + (eps * clip_adv_pattern)
    perturbed_img = tf.clip_by_value(perturbed_img, clip_value_min=0.0, clip_value_max=1.0)
    # Predict class after adversarial attack
    aa_pred_class, aa_pred_prob = get_model_preds(perturbed_img, model)

    # Visualize results
    fig, ax = plt.subplots(1, 3, figsize=(15, 4))
    ax[0].imshow(tf_img[0].numpy().squeeze(), cmap='gray')
    ax[0].set_title('Before Adversarial Attack\nPred Class: {}  Prob: {:.3f}'.format(ba_pred_class, ba_pred_prob))

    ax[1].imshow(clip_adv_pattern[0].numpy().squeeze(), cmap='gray')
    ax[1].set_title('Adversarial Pattern - EPS: {}'.format(eps))

    ax[2].imshow(perturbed_img[0].numpy().squeeze(), cmap='gray')
    ax[2].set_title('After Adversarial Attack\nPred Class: {}  Prob: {:.3f}'.format(aa_pred_class, aa_pred_prob))
    plt.show()

# Load your CIFAR-10 model (cnn) and define the classes array
# ...

# Sample index of a vehicle image from the test set
sample_vehicle_idx = 6
sample_vehicle_img = x_test[sample_vehicle_idx]
sample_vehicle_labelidx = y_test[sample_vehicle_idx]

# Perform FGSM attack on the sample vehicle image
perform_adversarial_attack_fgsm(input_image=sample_vehicle_img,
                                image_label_idx=sample_vehicle_labelidx,
                                model=model, loss_func=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                                eps=0.5)
```

# 1) Adversarial Training


 Train the model on adversarial examples generated using FGSM. By including adversarial examples during training, the model learns to be robust against these attacks.

```python
# Function to perform adversarial training with FGSM
def perform_adversarial_training(x_train, y_train, model, loss_func, eps=0.01, num_epochs=10, batch_size=32):
    for epoch in range(num_epochs):
        print(f"Epoch {epoch + 1}/{num_epochs}")
        for i in range(0, len(x_train), batch_size):
            x_batch = x_train[i:i + batch_size]
            y_batch = y_train[i:i + batch_size]

            # Regular training on clean examples
            model.train_on_batch(x_batch, y_batch)

            # Generate adversarial examples using FGSM
            with tf.GradientTape() as tape:
                tape.watch(x_batch)
                preds = model(x_batch)
                loss = loss_func(y_batch, preds)

            gradient = tape.gradient(loss, x_batch)
            signed_grad = tf.sign(gradient)

            adversarial_x_batch = x_batch + (eps * signed_grad)
            adversarial_x_batch = tf.clip_by_value(adversarial_x_batch, clip_value_min=0.0, clip_value_max=1.0)

            # Adversarial training on the generated adversarial examples
            model.train_on_batch(adversarial_x_batch, y_batch)

    print("Adversarial Training Completed.")
```

* Effect: This noise can disrupt the image's features and cause misclassification.

# Median Filtering Defence
 To defend against the Salt-and-Pepper Noise attack, we can use a technique called Median Filtering. Median Filtering is a simple yet effective method for removing impulse noise, such as salt-and-pepper noise, from an image.



```python
import scipy.ndimage

# Function to apply Median Filtering to an image
def apply_median_filter(image, filter_size=3):
    # Apply Median Filtering using scipy.ndimage.median_filter
    filtered_image = scipy.ndimage.median_filter(image, size=filter_size)
    return filtered_image

# Example of applying Median Filtering to a noisy image
filtered_image = apply_median_filter(np.copy(noisy_image), filter_size=3)

# Display the noisy and filtered images
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(sample_image)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(noisy_image)
plt.title("Noisy Image (Salt-and-Pepper)")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(filtered_image)
plt.title("Filtered Image (Median Filter)")
plt.axis('off')

plt.tight_layout()
plt.show()
```

We'll use metrics such as Peak Signal-to-Noise Ratio (PSNR) and Structural Similarity Index (SSIM) to quantitatively evaluate the denoising effect of Median Filtering.

```python
import skimage.metrics as metrics

# Function to calculate PSNR and SSIM between two images
def calculate_metrics(original_image, filtered_image):
    psnr_value = metrics.peak_signal_noise_ratio(original_image, filtered_image)
    ssim_value = metrics.structural_similarity(original_image, filtered_image, multichannel=True)
    return psnr_value, ssim_value

# List to store PSNR and SSIM values for each noisy image
psnr_values = []
ssim_values = []

# Perform Median Filtering and calculate metrics for all noisy images
for i in range(len(x_test)):
    noisy_image = apply_salt_and_pepper_noise(np.copy(x_test[i]), noise_level=0.1)
    filtered_image = apply_median_filter(np.copy(noisy_image), filter_size=3)
    psnr, ssim = calculate_metrics(x_test[i], filtered_image)
    psnr_values.append(psnr)
    ssim_values.append(ssim)

# Plot the PSNR values
plt.figure(figsize=(8, 4))
plt.plot(psnr_values, 'bo', label='PSNR')
plt.xlabel('Image Index')
plt.ylabel('PSNR Value')
plt.title('Peak Signal-to-Noise Ratio (PSNR) of Noisy Images')
plt.legend()
plt.grid(True)
plt.show()

# Plot the SSIM values
plt.figure(figsize=(8, 4))
plt.plot(ssim_values, 'ro', label='SSIM')
plt.xlabel('Image Index')
plt.ylabel('SSIM Value')
plt.title('Structural Similarity Index (SSIM) of Noisy Images')
plt.legend()
plt.grid(True)
plt.show()
```

# Normalization Defence
To defend against brightness adjustment attacks, we can use image normalization. Image normalization involves scaling the pixel values of the image to a certain range (e.g., [0, 1]) so that the image's brightness is consistent across different images. By normalizing the images, we can reduce the impact of brightness adjustments on the model's performance.

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to apply Brightness Adjustment to an image
def apply_brightness_adjustment(image, brightness_factor=0.9):
    noisy_image = np.clip(image + brightness_factor, 0.0, 1.0)
    return noisy_image

# Function to normalize an image
def normalize_image(image):
    min_val = np.min(image)
    max_val = np.max(image)
    normalized_image = (image - min_val) / (max_val - min_val)
    return normalized_image

# Example of applying Brightness Adjustment to a sample image
sample_image = x_test[9]  # Choose any image from the test set
noisy_image = apply_brightness_adjustment(np.copy(sample_image), brightness_factor=0.2)

# Normalize the noisy image
normalized_noisy_image = normalize_image(noisy_image)

# Display the original, noisy, and normalized images
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(sample_image)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(noisy_image)
plt.title("Noisy Image (Brightness Adjustment)")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(normalized_noisy_image)
plt.title("Normalized Noisy Image")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Function to generate adversarial pattern with translation attack
def generate_adversarial_pattern_translation(input_image, image_label_idx, model, loss_func, eps_x=5, eps_y=5):
    with tf.GradientTape() as tape:
        tape.watch(input_image)
        preds = model(input_image)
        loss = loss_func(image_label_idx, preds)

    gradient_x = tape.gradient(loss, input_image, dy=None, dx=eps_x)
    gradient_y = tape.gradient(loss, input_image, dy=eps_y, dx=None)

    translated_grad = tf.concat([gradient_x, gradient_y], axis=0)
    signed_grad = tf.sign(translated_grad)
    return signed_grad

# Perform FGSM attack with translation on the sample vehicle image
def perform_adversarial_attack_translation(input_image, image_label_idx, model, loss_func, eps_x=5, eps_y=5):
    tf_img = tf.convert_to_tensor(input_image)
    # Predict class before adversarial attack
    ba_pred_class, ba_pred_prob = get_model_preds(tf_img, model)
    # Generate adversarial image with translation
    adv_pattern = generate_adversarial_pattern_translation(tf_img, image_label_idx, model, loss_func, eps_x, eps_y)
    clip_adv_pattern = tf.clip_by_value(adv_pattern, clip_value_min=-1, clip_value_max=1)

    perturbed_img = tf_img + clip_adv_pattern
    perturbed_img = tf.clip_by_value(perturbed_img, clip_value_min=0.0, clip_value_max=1.0)
    # Predict class after adversarial attack
    aa_pred_class, aa_pred_prob = get_model_preds(perturbed_img, model)

    # Visualize results
    fig, ax = plt.subplots(1, 3, figsize=(15, 4))
    ax[0].imshow(tf_img[0].numpy().squeeze(), cmap='gray')
    ax[0].set_title('Before Adversarial Attack\nPred Class: {}  Prob: {:.3f}'.format(ba_pred_class, ba_pred_prob))

    ax[1].imshow(clip_adv_pattern.numpy().squeeze(), cmap='gray')
    ax[1].set_title('Adversarial Pattern - Translation')

    ax[2].imshow(perturbed_img[0].numpy().squeeze(), cmap='gray')
    ax[2].set_title('After Adversarial Attack\nPred Class: {}  Prob: {:.3f}'.format(aa_pred_class, aa_pred_prob))
    plt.show()
```

* Description: Randomly crop a portion of the input image.
* Effect: Cropping can remove critical information from the image and lead to misclassification.

# SimBA (Simple Black-Box Attack):

SimBA is an iterative black-box attack that performs random perturbations to the input image. It then accepts or rejects each perturbation based on its effect on the model's prediction.

In this code, we define the apply_simba_attack() function to apply the SimBA attack to an input image. The function generates random perturbations and applies them to the image. If a perturbation causes a change in prediction, it rejects the perturbation and retains the original image.

We then apply the SimBA attack to a sample image from the test set and display the original and perturbed images side by side.

SimBA is a simple yet effective black-box attack that can create adversarial examples with relatively low computational cost. It demonstrates the vulnerability of machine learning models to exploratory attacks that iteratively search for adversarial perturbations without access to the model's internals.

```python
import numpy as np

# Function to apply SimBA attack to an image
def apply_simba_attack(model, image, epsilon=0.1, num_iters=1000):
    original_pred = np.argmax(model.predict(np.expand_dims(image, axis=0)), axis=1)
    perturbed_image = np.copy(image)

    for _ in range(num_iters):
        # Generate random perturbation
        perturbation = np.random.uniform(-epsilon, epsilon, size=image.shape)
        perturbed_image = np.clip(perturbed_image + perturbation, 0.0, 1.0)

        # Check if perturbation decreases the confidence of the original prediction
        perturbed_pred = np.argmax(model.predict(np.expand_dims(perturbed_image, axis=0)), axis=1)
        if perturbed_pred != original_pred:
            # Reject the perturbation if it leads to a different prediction
            perturbed_image = np.copy(image)

    return perturbed_image

# Example of applying SimBA attack to a sample image
sample_image = x_test[7]  # Choose any image from the test set
perturbed_image_simba = apply_simba_attack(model, np.copy(sample_image), epsilon=0.1, num_iters=1000)

# Display the original and perturbed images
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(sample_image)
plt.title("Original Image")
plt.axis('on')

plt.subplot(1, 2, 2)
plt.imshow(perturbed_image_simba)
plt.title("Perturbed Image (SimBA)")
plt.axis('off')

plt.tight_layout()
plt.show()
```

To provide a graphical plot for the SimBA attack, we can demonstrate how the perturbed image changes over multiple iterations of the attack. This will show how the perturbation gradually evolves to fool the model. We'll create a grid of images, each corresponding to the image after a certain number of iterations.

```python
# Function to apply SimBA attack iteratively and store perturbed images at each iteration
def apply_simba_attack_iterative(model, image, epsilon=0.1, num_iters=1000):
    original_pred = np.argmax(model.predict(np.expand_dims(image, axis=0)), axis=1)
    perturbed_images = [np.copy(image)]

    for _ in range(num_iters):
        # Generate random perturbation
        perturbation = np.random.uniform(-epsilon, epsilon, size=image.shape)
        perturbed_image = np.clip(perturbed_images[-1] + perturbation, 0.0, 1.0)

        # Check if perturbation decreases the confidence of the original prediction
        perturbed_pred = np.argmax(model.predict(np.expand_dims(perturbed_image, axis=0)), axis=1)
        if perturbed_pred != original_pred:
            # Reject the perturbation if it leads to a different prediction
            perturbed_image = np.copy(perturbed_images[-1])

        perturbed_images.append(perturbed_image)

    return perturbed_images

# Apply SimBA attack iteratively to a sample image
sample_image = x_test[7]  # Choose any image from the test set
perturbed_images_simba = apply_simba_attack_iterative(model, np.copy(sample_image), epsilon=0.1, num_iters=100)

# Display the original and perturbed images at different iterations
num_cols = 10
num_rows = (len(perturbed_images_simba) - 1) // num_cols + 1

plt.figure(figsize=(15, 5 * num_rows))
for i, perturbed_image in enumerate(perturbed_images_simba):
    plt.subplot(num_rows, num_cols, i + 1)
    plt.imshow(perturbed_image)
    plt.title(f"Iteration {i}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

# Input Preprocessing Defence
To defend against the SimBA attack, we can apply a simple but effective defense called "Input Preprocessing." In this defense, we will preprocess the input image before feeding it to the model for prediction. Specifically, we will add a small amount of random noise to the image to make it robust against adversarial perturbations. The idea is to use this random noise as a form of regularization to make it difficult for the attacker to craft effective perturbations.

```python
import numpy as np

# Function to apply Input Preprocessing (Random Noise) to an image
def apply_input_preprocessing(image, noise_factor=0.01):
    noise = np.random.uniform(-noise_factor, noise_factor, size=image.shape)
    perturbed_image = np.clip(image + noise, 0.0, 1.0)
    return perturbed_image

# Function to apply SimBA attack to an image
def apply_simba_attack(model, image, epsilon=0.1, num_iters=1000):
    original_pred = np.argmax(model.predict(np.expand_dims(image, axis=0)), axis=1)
    perturbed_image = np.copy(image)

    for _ in range(num_iters):
        # Generate random perturbation
        perturbation = np.random.uniform(-epsilon, epsilon, size=image.shape)
        perturbed_image = np.clip(perturbed_image + perturbation, 0.0, 1.0)

        # Check if perturbation decreases the confidence of the original prediction
        perturbed_pred = np.argmax(model.predict(np.expand_dims(perturbed_image, axis=0)), axis=1)
        if perturbed_pred != original_pred:
            # Reject the perturbation if it leads to a different prediction
            perturbed_image = np.copy(image)

    return perturbed_image

# Step 1: Choose a sample image from the test set
sample_image = x_test[7]  # Choose any image from the test set

# Step 2: Apply SimBA Attack to the Sample Image
perturbed_image_simba = apply_simba_attack(model, np.copy(sample_image), epsilon=0.1, num_iters=1000)

# Step 3: Apply Input Preprocessing Defense to the Perturbed Image (SimBA Attack)
defensed_image = apply_input_preprocessing(np.copy(perturbed_image_simba), noise_factor=0.01)

# Step 4: Display the Original, Perturbed, and Defensed Images
plt.figure(figsize=(12, 4))

# Original Image
plt.subplot(1, 3, 1)
plt.imshow(sample_image)
plt.title("Original Image")
plt.axis('on')

# Perturbed Image (SimBA Attack)
plt.subplot(1, 3, 2)
plt.imshow(perturbed_image_simba)
plt.title("Perturbed Image (SimBA Attack)")
plt.axis('off')

# Defensed Image (Input Preprocessing)
plt.subplot(1, 3, 3)
plt.imshow(defensed_image)
plt.title("Defensed Image (Input Preprocessing)")
plt.axis('off')

plt.tight_layout()
plt.show()
```
